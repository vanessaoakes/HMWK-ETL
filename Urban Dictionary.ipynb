# Cleaning:

Finding Duplicates
-the column rows were not adding up so I needed to remove several "NAN" to continue

```python
#Finding Duplicates
dupe_defs = urban_dictionary_df.duplicated(subset='definition', keep=False)
urban_dictionary_df[dupe_defs].sort_values(by='definition')
```

```python
#How many row did we start with?
df = urban_dictionary_df
df.count()
```

```python
#Remove null word_id
df = df[df.word_id.notnull()]
df.count()
```

```python
#Remove word 
df = df[df.word.notnull()]
df.count()
```

```python
#Remove null definitions
df = df[df.definition.notnull()]
df.count()
```

```python
#Remove null authors
df = df[df.author.notnull()]
df.count()
```

```python
#Final row count:
df.count()
```

Author most contribute:

```python
#Authored most contributed
authors_contributed=df.groupby("author")
authors_contributed["author"].count().sort_values(ascending=False).head()
```

# Connect to local database

```python
urban_dictionary_string = "urban_dictionary_db"
engine = create_engine(f'sqlite:///{urban_dictionary_string}')

if not database_exists(engine.url):
    create_database(engine.url)

print(database_exists(engine.url))
```

```python
# Check for tables
```

```python
engine.table_names()
```

```python
# Use pandas to load csv converted DataFrame into database
```

```python
df.to_sql(name='Urban_D', con=engine, if_exists='append', index=False, index_label="word_id")
```
